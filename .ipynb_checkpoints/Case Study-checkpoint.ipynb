### Start extracting Data from Internet itself using the foloowing steps as follows:

```python



import requests            
import json
import matplotlib.pyplot as plt
a=requests.get('https://www.quandl.com/api/v3/datasets/NSE/SBIN.json?api_key=udbBDns7sGhKiRa7yn1W')  #url of the File which you want to read
b=json.loads(a.text) #loads the data into another String
my_data=b['dataset']['data']  #this will load the data into the my_data
useful_data=[i[1] for i in reversed(my_data)]
x_data=[i for i in range(len(useful_data))]
plt.plot(x_data,useful_data)
plt.show()
```
